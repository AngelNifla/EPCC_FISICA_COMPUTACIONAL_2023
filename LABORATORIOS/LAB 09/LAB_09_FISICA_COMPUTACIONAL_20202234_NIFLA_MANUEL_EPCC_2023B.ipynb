###***Funciones Principales :***

* Se define la función "Regla" para la Regla de Selección:


```python
def Regla(C_Left, C_Current, C_Rigth):
    # Regla de selección
    if C_Left == 0 and C_Current == 0 and C_Rigth == 0:
        return 0
    elif C_Left == 0 and C_Current == 0 and C_Rigth == 1:
        return 1
    elif C_Left == 0 and C_Current == 1 and C_Rigth == 0:
        return 1
    elif C_Left == 0 and C_Current == 1 and C_Rigth == 1:
        return 1
    elif C_Left == 1 and C_Current == 0 and C_Rigth == 0:
        return 1
    elif C_Left == 1 and C_Current == 0 and C_Rigth == 1:
        return 0
    elif C_Left == 1 and C_Current == 1 and C_Rigth == 0:
        return 0
    elif C_Left == 1 and C_Current == 1 and C_Rigth == 1:
        return 0
```

* Se define la funcion "Evolucion" para la Evolución del Automata:

```python
def Evolucion(Initial_State, num):
    automata = [Initial_State]

    for paso in range(num):
        estado_actual = automata[-1]
        nuevo_estado = [0]  #Celda ficticia

        for i in range(len(estado_actual) - 2):
            C_Left = estado_actual[i]
            C_Current = estado_actual[i + 1]
            C_Rigth = estado_actual[i + 2]
            nueva_celda = Regla(C_Left, C_Current, C_Rigth)
            nuevo_estado.append(nueva_celda)

        nuevo_estado.append(0)  #Celda ficticia
        automata.append(nuevo_estado)

    return automata
```

###***Inicio de Ejecución :***

```python
Initial_State = []
n = int(input("Numero de datos para el Estado inicial: "))
for i in range(n):
  Initial_State.append(int(input("\tDato: ")))

num = int(input("Numero de Pasos te tiempo: "))

automata_resultante = Evolucion(Initial_State, num)
print(f"\n\n")
for paso, estado in enumerate(automata_resultante):
    print(f"Paso {paso}: {estado[1:-1]}")
```
