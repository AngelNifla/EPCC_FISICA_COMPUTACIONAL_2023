#### **Ejercicios:**

Resuelva las siguientes integrales por método Monte Carlo.

*   Calcule $I[g(X)] = \int_{0}^{1} e^{x^2} \, dx$.

*   Calcule $I[g(X)] = \int_{-1}^{1} e^x4 \, dx$.
*   Calcule $I[g(X)] = \int_{0}^{1} (1-e^{x^2})^{\frac{1}{2}} \, dx$.
*   Calcule $I[g(X)] = \int_{0}^{\infty} x(1+x^2)^{-2} \, dx$.
*   Calcule $I[g(X)] = \int_{0}^{1} e^{x+x^{2}}\, dx$.
*   Calcule $I[g(X)] = \int_{0}^{\infty} e^{-x} \, dx$.
*   Calcule $I[g(X)] = \int_{0}^{\infty} (1-x^2)^{\frac{3}{2}} \, dx$.

###***Hallando integrales :***




```python
import sympy as sp

# Definir la variable simbólica
x = sp.symbols('x')

# Definir las funciones g(X) para cada integral

g1 = sp.exp(x**2)
g2 = sp.exp(x)*4
g3 = sp.sqrt(1 - sp.exp(x**2))
g4 = x / (1 + x**2)**2
g5 = sp.exp(x + x**2)
g6 = sp.exp(-x)
g7 = (1 - x**2)**(3/2)

# Calcular las integrales analíticamente

result1 = sp.integrate(g1, (x, 0, 1))
result2 = sp.integrate(g2, (x, -1, 1))
result3 = sp.integrate(g3, (x, 0, 1))
result4 = sp.integrate(g4, (x, 0, sp.oo))
result5 = sp.integrate(g5, (x, 0, 1))
result6 = sp.integrate(g6, (x, 0, sp.oo))
result7 = sp.integrate(g7, (x, 0, 1))

# Mostrar los resultados

print("Resultado 1:", result1.evalf())
print("Resultado 2:", result2.evalf())
print("Resultado 3:", result3.evalf())
print("Resultado 4:", result4.evalf())
print("Resultado 5:", result5.evalf())
print("Resultado 6:", result6.evalf())
print("Resultado 7:", result7.evalf())
```

###***Hallando integrales por Monte Carlo :***




```python
import numpy as np
import sympy as sp

# Definir la función g(X) para cada integral

def g1(x):
    return np.exp(x**2)

def g2(x):
    return np.exp(x**4)

def g3(x):
    return np.sqrt(1 - np.exp(x**2))

def g4(x):
    return x / (1 + x**2)**2

def g5(x):
    return np.exp(x + x**2)

def g6(x):
    return np.exp(-x)

def g7(x):
    return (1 - x**2)**(3/2)

# Función para realizar la integración por método Monte Carlo

def monte_carlo_integration(func, a, b, num_samples=10000):
    x_values = np.random.uniform(a, b, num_samples)
    y_values = func(x_values)
    integral_estimate = (b - a) * np.mean(y_values)
    return integral_estimate

# Calcular las integrales por método Monte Carlo

result1 = monte_carlo_integration(g1, 0, 1)
result2 = monte_carlo_integration(g2, -1, 1)
result3 = monte_carlo_integration(g3, 0, 1)
#result4 = monte_carlo_integration(g4, 0, sp.oo)
result5 = monte_carlo_integration(g5, 0, 1)
#result6 = monte_carlo_integration(g6, 0, sp.oo)
result7 = monte_carlo_integration(g7, 0, 1)

# Mostrar los resultados

print("Resultado 1:", result1)
print("Resultado 2:", result2)
print("Resultado 3:", result3)
print("Resultado 4:", result4)
print("Resultado 5:", result5)
print("Resultado 6:", result6)
print("Resultado 7:", result7)
```
