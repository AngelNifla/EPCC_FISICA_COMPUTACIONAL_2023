```python
import matplotlib.pyplot as plt
import numpy as np
import cmath

def newton(function, derivative, x0, tol, max_iter=100):
    x1 = 0
    a=0
    if abs(x0-x1)<= tol and abs((x0-x1)/x0)<= tol:
        return x0

    #print("k\t x0\t\t function(x0)")
    k = 1
    while k <= max_iter:
        f = function(x0)
        d = derivative(x0)
        if d!=0.0:
          x1 = x0 - (f/d)
          #print("x%d\t%e\t%e"%(k,abs(x1),abs(function(x1))))

          if abs(x0-x1)<= tol and abs((x0-x1)/x0)<= tol:
            plt.plot(x0, function(x0), 'or')
            return x1

          x0 = x1
          k = k + 1
          plt.plot(x0, function(x0), 'or')

          # Stops the method if it hits the number of maximum iterations
          if k > max_iter:
              print("AVISO: Se superó el número máximo de iteraciones.")
        else:
          a=1
          k = k + 1
    print("Calculando...")
    if a:
      print("No se puede dividir por cero...")
    return x1  # Returns the value

def graf():
  plt.plot(u, w)
  plt.axhline(y=0.0, color='black', linestyle='-')
  plt.title('Newton-Raphson Graphics ')
  plt.xlabel('X')
  plt.ylabel('Y')
  plt.grid(True)
  plt.legend(['Xn'], loc='upper left')
  plt.show()
```

<center>$\LARGE{y = ln(x-2)}$</center>


\
*   La función es : $f(x) = ln(x-2)$
*   La derivada de $f(x)$ es $f'(x) = \frac{1}{x-2}$



```python
def function(x):
    value = (cmath.log(x-2))
    return value #function

def derivative(x):
    return (1/x-2)  #derivative

sqrt = newton(function, derivative, 3, 0.0000001)
print("El valor aproximado de x es: "+str(sqrt))

u = np.arange(1.0, 2.0, 0.0001)
w = [cmath.log(x - 2) for x in u]
graf()
```

*   La función es : $f(x) = e^{-x}$

```python
def function(x):
    value = (cmath.e**(-x))
    return value #function

def derivative(x):
    return (-cmath.e**x)-(x)  #derivative

sqrt = newton(function, derivative, 4, 0.0000001)
print("El valor aproximado de x es: "+str(sqrt))

u = np.arange(1.0, 2.0, 0.0001)
w = (cmath.e**(-u))
graf()
```

\
*   La función es : $f(x) = (x-2)^{\frac{1}{2}}$
