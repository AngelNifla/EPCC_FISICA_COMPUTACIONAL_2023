Las magnitudes de $∆x$, $Vi$ y $∆t$  escalarmente no podría presentarse como magnitudes negativas, es decir, no existe una distancia o velocidad o variación de tiempo negativa. Por otro lado la magnitud $a$ si puede llegar a presentar valores negativos, esto debido a que es una magnitud que escalarmente interpreta la aceleración y desaceleración de un cuerpo.

2. Específicas: \\


*   Para poder hallar la $Vi$ y $a$ se debe despejar estas en la ecuación, obteniendo $Vi =\frac{2∆x-a∆t^2}{2∆t}$ y $ a =\frac{2(∆x-Vi∆t)}{∆t^2}$ respectivamente; donde se descrubre que en ambos despejes $∆t \neq 0$.
