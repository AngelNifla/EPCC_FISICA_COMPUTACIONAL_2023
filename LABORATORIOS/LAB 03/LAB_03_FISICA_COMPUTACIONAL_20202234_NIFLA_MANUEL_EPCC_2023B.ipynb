### $SOLUCION \ EJERCICIO $


##***ANALISIS:***

La ecuación de la tercera Ley de Newton es
<center> $F = ma$ </center>
Donde:

*   **$F:$** es la fuerza.
*   **$m:$** es la masa.
*   **$a:$** es la aceleración.

<center><img src="https://www.ecured.cu/images/a/a9/Acci%C3%B3n_y_reacci%C3%B3n_.jpeg" width="250"></center>


**Condiciones**

*   El valor de $m$ debe ser positiva.

En conclusión, la implementación del código permite hallar la tercera ley de Newton que toma la masa y la aceleración como argumentos y devuelve la fuerza calculada.

##***CODIGO:***

```python
def calcular_fuerza(m, a):
    f = m * a
    return f

def calcular_masa(f, a):
    m = f / a
    return m

def calcular_aceleracion(f, m):
    a = f / m
    return a

opcion = int(input("¿Qué deseas calcular?\n1. fuerza\n2. masa\n3. aceleracion\n"))

if opcion == 1:
    masa = float(input("Ingresa la masa: "))
    while masa < 0:
      masa = float(input("Ingresa la masa (positivo): "))
    aceleracion = float(input("Ingresa la aceleracion: "))
    resultado = calcular_fuerza(masa, aceleracion)
    print(f"-> La fuerza es: {resultado} Newtons")

elif opcion == 2:
    fuerza = float(input("Ingresa la fuerza: "))
    aceleracion = float(input("Ingresa la aceleracion: "))
    resultado = calcular_masa(fuerza, aceleracion)
    print(f"-> La masa es: {resultado} kilogramos")

elif opcion == 3:
    fuerza = float(input("Ingresa la fuerza: "))
    masa = float(input("Ingresa la masa: "))
    while masa < 0:
      masa = float(input("Ingresa la masa (positivo): "))
    resultado = calcular_aceleracion(fuerza, masa)
    print(f"-> La aceleracion es: {resultado} metros por segundo al cuadrado")

else:
    print("Opción no válida. Por favor, elige entre fuerza, masa o aceleracion.")

```
