##***ANALISIS:***

La energía mecánica total de un sistema puede descomponerse en: la energía cinética y la energía potencial. Estas dos formas de energía están relacionadas y cambian a medida que el sistema se mueve.

Descripción:

*   **Energía Cinética ($E_c$):** La energía cinética está asociada al movimiento de un objeto.<center> $E_c = \frac{1}{2}mv^2$ </center>
\
Donde $m$ es la masa del objeto y $v$ es su velocidad.

*   **Energía Potencial ($E_p$):** La energía potencial está asociada a la posición de un objeto en un campo de fuerza, como el campo gravitatorio.<center> $E_p = mgh$ </center>
\
Donde $m$ es la masa, $g$  es la aceleración debida a la gravedad y $h$ es la altura sobre un punto de referencia.
*   **Energía Mecánica Total ($E_m$):** La energía mecánica total es la suma de la energía cinética y la energía potencial.<center> $E_m = E_c + E_p$ </center>

\
<center><img src="https://s5.static.brasilescola.uol.com.br/be/2020/03/1-energia-mecanica.jpg" width="400"></center>

\
*   **Evolución con respecto al Tiempo ($E_m(t)$):** Con las ecuaciones de movimiento del cuerpo.<center> $v = - gt$
\
$h = h_0 - \frac{1}{2}gt^2$</center>Se tiene:<center> $ E_c(t) = \frac{1}{2}m(v^2) = \frac{1}{2}mg^2t^2$
\
\
$ E_p(t) = mg(h) = mgh_0-\frac{1}{2}mg^2t^2 = mgh_0-E_c(t)$</center>Entonces:<center> **$E_m(t) = mgh_0$** </center>

**Condiciones**

*   En las fórmulas $m$, $g$, $h$ y $v$ deben ser positivas.

\
En conclusión, la implementación del código permite hallar y graficar la evolución de las Energías Cinética, Potencial y Mecánica con respecto del tiempo, permitiendo solo el ingreso de datos positivos para la $m$, $g$, $h$ y $v$, caso contrario volverá a pedir nuevos datos para así cumplir con las condiciones.

##***CODIGO:***

```python
import matplotlib.pyplot as plt
import numpy as np

########## Funciones para la energía cinética, potencial y mecánica ############

def Ec(m, v):
    return 0.5 * m * v**2

def Ep(m, g, h):
    return m * g * h

def Em(c, p):
    return c + p

################## Parámetros #################

masa = float(input("Ingrese la masa del móvil (en kg): "))
while masa<0:
  masa = float(input("Ingrese la masa del móvil (en kg, positivo): "))
gravedad = float(input("Ingrese la aceleracion de la gravedad (en m/s^2): "))
while gravedad<0:
  gravedad = float(input("Ingrese la aceleracion de la gravedad (en m/s^2, positivo): "))
altura = float(input("Ingrese la altura (en m): "))
while altura<0:
  altura = float(input("Ingrese la altura (en m, positivo): "))
velocidad_ini = float(input("Ingrese la velocidad inicial (en m/s): "))
while velocidad_ini<0:
  velocidad_ini = float(input("Ingrese la velocidad inicial (en m/s, positivo): "))


######## Calculando las energías en función del tiempo ########
tiempo = np.linspace(0, 5, 100)

Altura = altura - 0.5 * gravedad * tiempo**2
velocity = -gravedad * tiempo
E_cinetica = Ec(masa, velocity)
E_potencial = Ep(masa, gravedad, Altura)
E_mecanica = Em(E_cinetica, E_potencial)

#################  Gráfico #############################
plt.figure(figsize=(10, 6))

plt.plot(tiempo, E_cinetica, label='Energía Cinética')
plt.plot(tiempo, E_potencial, label='Energía Potencial')
plt.plot(tiempo, E_mecanica, label='Energía Mecánica')

plt.title('Energías en función del tiempo')
plt.xlabel('Tiempo (segundos)')
plt.ylabel('Energía (Joules)')
plt.legend()
plt.grid(True)
plt.show()


```
